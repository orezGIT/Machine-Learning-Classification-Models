# Data Preprocessing 


```python
#counting the number of instances in the target column
df["Churn"].value_counts()
```

```python
sns.countplot(df, x="Churn", hue="Churn")
plt.show()
```

```python
#filter the relevant columns
x = df.drop("Churn", axis=1) 
y = df["Churn"] 
```

```python
# Split data into train and test sets 
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0, stratify=y)
```

```python
#Apply class balancing 
#use smote method
from imblearn.over_sampling import SMOTE 
resampler = SMOTE(random_state=0) 
x_train_smote, y_train_smote = resampler.fit_resample(x_train, y_train) 

#visualise the balanced class
sns.countplot(x=y_train_smote, hue = y_train_smote) 
plt.show() 
```

```python
# Apply standard scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_s = scaler.fit_transform(x_train_smote)
X_test_s = scaler.transform(x_test)
```

# Model Training 

```python
#fitting the K-NN to the training set
from sklearn.neighbors import KNeighborsClassifier 
classifier = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2) 
classifier.fit(X_train_s, y_train_smote)
```

```python
#predicting the test set results
y_predict = classifier.predict(X_test_s) 
```

# Model Evaluation 

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, confusion_matrix
print(f"Accuracy Score: {accuracy_score(y_test, y_predict) * 100:.2f}%") 
cm = metrics.confusion_matrix(y_test, y_predict) 
print("Confusion Matrix:")
print(cm, "\n\n") 
print("-------------------------------------------------------------------------")
result = metrics.classification_report(y_test, y_predict)
print("Classification Report:\n") 
print(result)
```

```python
#visualising the confusion matrix 
cm = confusion_matrix(y_test,y_predict) 
ax = sns.heatmap(cm, cmap='flare',annot=True, fmt='d') 
plt.xlabel("Predicted Class",fontsize=12) 
plt.ylabel("True Class",fontsize=12) 
plt.title("Confusion Matrix",fontsize=12) 
plt.show()
```

# Decision Tree Model

```python
#Note: this is basically for creating and training the Decision Tree model
#fitting the K-
from sklearn.tree import DecisionTreeClassifier  
classifier =   DecisionTreeClassifier(criterion="entropy", random_state=0) 
classifier.fit(x_train_smote, y_train_smote)
```

```python
#predicting the test set results
y_pred_tree = classifier.predict(x_test) 
```

# Model Evaluation

```python
from sklearn import metrics 
print(f"Accuracy Score: {accuracy_score(y_test, y_pred_tree)*100:.2f}%") 
cm1 = metrics.confusion_matrix(y_test, y_pred_tree) 
print("Confusion Matrix:")
print(cm1, "\n\n") 
print("-------------------------------------------------------------------------")
result = metrics.classification_report(y_test,y_pred_tree)
print("Classification Report:\n") 
print(result)
```

```python
#visualising the confusion matrix 
cm1 = confusion_matrix(y_test,y_pred_tree) 
ax = sns.heatmap(cm1, cmap='flare',annot=True, fmt='d') 
plt.xlabel("Predicted Class",fontsize=12) 
plt.ylabel("True Class",fontsize=12) 
plt.title("Confusion Matrix",fontsize=12) 
plt.show()
```
